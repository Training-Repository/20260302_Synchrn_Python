```python
# OPTIONS operations

# Create the request object
import urllib.request
req = urllib.request.Request(bin_url, method='OPTIONS')

# Send the request
resp = urllib.request.urlopen(req)

print(resp.info())
```

```python
# POST operation

# Prepare data to post
import urllib.parse
import urllib.request
post_data = urllib.parse.urlencode({"name": "Amit", "city": "Pune"}).encode('ascii')

# Prepare the request
req = urllib.request.Request('https://httpbin.org/post', 
                             method='POST',
                             data=post_data,
                             headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"})

# Send the request
resp = urllib.request.urlopen(req)

print(resp.info())
print(f"\n\n{resp.status}, {resp.reason}")
```

```python
data = resp.read().decode('UTF-8')
pprint(data)
```

```python
# Exception handling for forbidden (403) response
from urllib.error import HTTPError
try:
    with urllib.request.urlopen('https://httpbin.org/status/403') as resp:
        print("Read response -->", resp.read().decode('UTF-8'), resp.status, resp.reason, sep="\n\n")
except HTTPError as ex:
    print(f"{ex!r}")
    print(f"Code: {ex.code}")
    print(f"Reason: {ex.reason}")
print("\n\nDone")
```

```python
# Exception handling for URL error
from urllib.error import URLError
from urllib.error import HTTPError
try:
    with urllib.request.urlopen('https://fake-site.com') as resp:
        print("Read response -->", resp.read().decode('UTF-8'), resp.status, resp.reason, sep="\n\n")
except (HTTPError, URLError) as ex:
    print(f"{ex!r}")
    # print(f"Code: {ex.code}")     # URLError doesn't have an attribute called 'code', as it never reached the server
    print(f"Reason: {ex.reason}")
print("\n\nDone")
```

```python
wikipedia_search_url = 'https://en.wikipedia.org/w/index.php'
params = {"search": "requests"}

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept-Language": "en-IN,en-GB;q=0.9,en-US;q=0.8,en;q=0.7"
}

resp = requests.get(wikipedia_search_url,
                    params=params,
                    headers=headers,
                    timeout=15)
print(resp.status_code)
print(resp.url)
```

```python
def print_response_history(resp):
    if resp.history:
        print("History")
        for r in resp.history:
            print(r.status_code, r.url)

        print("Finally")
        print(resp.status_code, resp.url)
    else:
        print("Request was not redirected.")
```

```python
print_response_history(resp)
```

```python
resp = requests.get('https://httpbin.org/absolute-redirect/5', allow_redirects=False)
print_response_history(resp)
```
